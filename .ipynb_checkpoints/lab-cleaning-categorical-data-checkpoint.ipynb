### 1. Import the necessary libraries if you are starting a new notebook.

````python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
````

### 2. Load the csv.

````python
customer_df = pd.read_csv("we_fn_use_c_marketing_customer_value_analysis.csv")
````

````python
customer_df.head(1)
````

I customise the column names.

````python
customer_df.columns = [col_name.lower().replace(' ', '_') for col_name in customer_df.columns]
````

````python
customer_df.rename(columns={'employmentstatus': 'employment_status'}, inplace=True)
customer_df.columns
````

And I again change the column 'effective_to_date' to datetime since it contains only dates and isn't exactyl providing categorical values.

````python
customer_df['effective_to_date'] = pd.to_datetime(customer_df['effective_to_date'])
print(customer_df['effective_to_date'].dtype)
````

### 3. What should we do with the customer_id column?

````python
customer_df.info()
````

The customer ID can be set as the index since it is a unique identifier of each row. This way it won't be a problem when encoding the categorical columns.

````python
customer_df = customer_df.set_index('customer')
````

### 4. Load the continuous and discrete variables into numerical_df and categorical_df variables.

````python
numerical_df = customer_df.select_dtypes(np.number)
categorical_df = customer_df.select_dtypes(np.object)
````

### 5. Plot every categorical variable. What can you see in the plots? 
#### Note that in the previous lab you used a bar plot to plot categorical data, with each unique category in the column on the x-axis and an appropriate measure on the y-axis. This time for each categorical variable you will have the unique categories in the column on the x-axis and the target(which is numerical) on the y-axis.

````python
categorical_df.columns
````

Both a barplot and a boxplot are adequate types to represent the relation of each of the categorical features with the target 'total_claim_amount'. I try them both for the relation of state/target.

````python
plt.figure(figsize=(12, 6))
sns.barplot(x='state', y='total_claim_amount', data=customer_df, palette='Blues', edgecolor='black')
plt.show()
````

````python
plt.figure(figsize=(12, 6))
sns.boxplot(x='state', y='total_claim_amount', data=customer_df, palette='Blues')
plt.show()
````

The boxplot shows me also the outliers for each unique category. But I wnat to focus on the majority, therefore I am using barplots to display all relations as following.

````python
# I create a matrix of the required size.
fig, axs = plt.subplots(7, 2, figsize=(17, 17)) 
axs = axs.flatten()

# I iterate through the different categorical columns, add a color scheme and set a title for each histogram.
for i, column in enumerate(categorical_df.columns):
    sns.barplot(x=column, y='total_claim_amount', data=customer_df, palette='Blues', edgecolor='black', ax=axs[i])
    axs[i].set_title(column)

# I improve the layout, because the plots were displayed too close to each other.
plt.tight_layout()
plt.show()
````

To summarize a few key findings:

- The total claim amount across the different locations is almost the same.
- The more expensive and better the coverage package, the higher the total claim amount. Customers with the premium coverage package are claiming the most.
- Customers with a lower education level claim more. Unemployed customers as well. 
- The total amount claimed in suburban and urban areas is abour 9 times higher than in rural areas.
- The renew offer 1 results in a higher total claim amount than offers 2-4.
- Customers with a luxurious car claim by far the most. The ratio is around 2:1.
- Gender, response rate, policy type and policy as well as the sales channel do not have an impact on the total amount claimed. 

### 6. For the categorical data, check if there is any data cleaning that needs to be performed. 
#### **Hint**: You can use the function 'value_counts()' on each of the categorical columns and check the representation of different categories in each column. Discuss if this information might in some way be used for data cleaning.

````python
# I apply value_counts to each categorical variable.
categorical_df['policy'].value_counts()
````

Using 'value_counts', I can observe possible high cardinality. For the variables in the categorical_df there isn't any. Solemly in the column 'policy' the categories could be bucketed into ['Personal', 'Corporate' and 'Special'] since there isn't much difference between the three of each and in the relation of all of the 9 categories with the target. Something I can also observe from the above barplots.
There are also no NaNs or null values (something I had already checked for in the last lab).
If the counts are also representative of the actual customer data, then there is no data cleaning required.
